```python
#Importar librerias
import numpy as np
import matplotlib.pyplot as plt
import datetime
import time
```

```python
#Importar archivo y separarlo en columnas. El skip_header permite eliminar el encabezado
ano,mes,minas,area = np.genfromtxt("desminadohumanitario.csv", delimiter=",", skip_header=1, unpack=True)
```

```python
#Creacion de una lista vacia para guardar los datos de fechas
f=[]
#Obtener la longitud de la lista ano para asignar a las matrices
l=len(ano)
#Ciclo para generar la fecha producto de la union entre anos y meses en el primer dia
for i in range(0,l):
    
    a=datetime.date(int (ano[i]),int (mes[i]),1)
    f.append(a)

#Creacion de matrices para trabajar las gráficas
fe = np.zeros(len(f))   #Corresponde a la matriz de las fechas en un numero especifico
feo=np.zeros(len(f))    #Corresponde a la matriz de las fechas en un numero especifico pequeño para que pueda ser usado con las exponencianles

#Ciclo para realizar la conversion de fechas
for u in range(len(f)):
    b=f[u].timetuple() #Convierte los años, meses, dias en una tupla 
    fe[u]=time.mktime(b) #Unifica la tupla en un dato numerico unico en segundos
    a=time.mktime(b)/(10**9)#Divido en 10 a la 9 porque sino los valores del exponencial son muy grandes para realizar regresiones
    feo[u]=a

```
