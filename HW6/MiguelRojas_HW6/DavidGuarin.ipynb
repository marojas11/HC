```python
#Se define la funcion en el cual tendra como parametro el numero dado por la persona
def random_2normal(l):
    #Se inicializaron unas variables
    q=1
    posicion =0
    ejex = []
    #En este se creo un arreglo en el cual seran numeros del 0 a l uno al azar y tendran la cantidad de numeros
    #ingresados por el parametro
    w = np.random.normal(0.0, 1, l)
  #recorrido en el cual ira por el numero puesto en el parametro
    for leng in range(len(w)):
        
        numero=w[leng]
        #Aqui encontrara el numero mas cercano del w perteneciente al eje y por lo cual enviara la posicion del x en ella
        for mm in range(len(yyy)):
            resta= abs(numero-yyy[mm])
            if(resta<q):
                q=resta
                posicion=mm
        #aqui guardarara el arreglo para los numeros en el eje x
        ejex.append(xx[posicion])   
    return ejex

histo= random_2normal(10000)
hist(histo)
print histo
    



```
