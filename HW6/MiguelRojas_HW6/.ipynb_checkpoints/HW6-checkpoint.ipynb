```python
import matplotlib.pyplot as plt
import numpy as np
import csv as csv
```

```python
 #Matriz de ceros
a= np.zeros((5,5))
d=1
for i in range (0,5):
    x=i%2 #Si es impar la fila asigne de derecha a izquierda
    
    if x==0:
        for j in range (0,5):
            a[i,j]=d
            d=d+1
    
    else:
         for j in range (0,5):
            a[i,4-j]=d
            d=d+1
            
print a
```

```python
a=[] #Inicializar listas
v1=[] #Fila 1, coordenada x
v2=[] #Fila2, coordenada y
v3=[] #Fial 


var1,var2 = np.genfromtxt("visiblestars.csv", delimiter=",", unpack=True) #genfromtxt

with open("namedstars.csv", 'rb') as csvfile: #importar desde csv.reader
    spamreader = csv.reader(csvfile, delimiter=",", quotechar='|') 
    for row in spamreader: 
        a.append(row) #agregar a las filas
    for i in range (len(a)): 
        b=a[i] 
        for j in range (len(b)):
            if(j==0): 
                v1.append(b[j]) #agregar a las filas
            if(j==1): 
                v2.append(b[j]) #agregar a las filas
            if(j==2): 
                v3.append(b[j])#agregar a las filas
                
figure(figsize(16,10)) #Tamaño
xlim([0,350]) #Tamaño del eje
ylim([-100,100])#Tamaño del eje
xlabel(r'$DEC/\degree$',fontsize =16) #etiqueta del eje
ylabel(r'$RA/\degree$',fontsize =16)
plt.plot(var1,var2, color='black',ms=0.5,ls='*',marker='.') #graficas
plt.plot(v1,v2, color="red", ms=10, ls='*',marker='o')
for i in range(len(v3)): 
    x= v1[i] 
    y = v2[i] 
    n= v3[i] 
    
    plt.text(x,y,n, fontsize=10,bbox=dict(facecolor='white',alpha=0.7),verticalalignment='bottom',horizontalalignment='left')
```

```python

figure(figsize(8,8))#Tamaño
a0=27890
a1=0-526.3 #constantes
a2=3.410
a3=-0.007502
C=116
mm=125
D=4.2

Ev = a0 + a1*var1 + a2*(var1)**2 + a3 *(var1)**3 #Formulas
Evv=Ev+C*exp(-(var1-mm)**2/D)
title(r'CMS $\sqrt{s}=7 TeV, L=5.1 fb^{-1}  \sqrt{s}=8 TeV, L=5.3 fb^{-1}$',fontsize=15) #titulo
ylabel(r'$S/(S+B)$ Weighted Events / $1.5 $GeV ', fontsize=15)
xlabel(r'$m_{yy}$ GeV ', fontsize=15)
xlim([105,153])

var1,var2 = np.genfromtxt("higgs.csv", delimiter=",", unpack=True) #Importacion


plt.plot(var1,Ev,ls='--', c='red',label="Ev") #Graficas
plt.plot(var1,Evv, c='b', label="Ev*")
plt.scatter(var1,var2,color='black')
plt.errorbar(var1,var2, yerr=35, label="dispersion",ls='none', color='black', marker="o")


plt.legend(frameon=False)
plt.savefig("CMSHiggs.pdf", format='PDF')


```
