```python
#Llamar paquetes de numpy para funcion vectorizar y matplotlib para exportar la imagen
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Metodo que permite calcular los valores de la funcion dada
def funcion(z):
    y=z**5+z**2
    return y
#Metodo que permite calcular los valores de la derivada de la funcion anterior
def derivada(z):
    dy=5*(z**4)+2*z
    return dy
```

```python

#Metodo para la funcion de Newton-Raphson

def Newton(x, epsilon):
    root = x
    iteration = 0.
    i=abs(funcion(root))
    #Ciclo para hallar las raices
    while i >= epsilon:
        
        if derivada(root)==0:
            print "error"
            i=epsilon+1
        else:
            iteration += 1 
            root = root - funcion(root)/(derivada(root))
           #Limite en las iteraciones
            if iteration>=50:
                i=epsilon+1
            else:
                i=abs(funcion(root))
    return root
```

```python
#Funcion para vectorizar la funcion de NewtonRaphson
funcionVectorizada=np.vectorize(Newton)
```

```python
#Creacion de una matriz de 200x200 con numeros imaginarios entre -1+j y 1-j
m=[] #Lista para las filas
reales=np.linspace(-1,1,200) #Lista de 200 Numeros entre -1 y 1
for i in range(0,200):
    b=[] #Lista para las columnas\n",
    imaginarios=np.linspace(1,-1,200) #Lista de 200 Numeros entre -1 y 1
    for k in range(0,200):
        b.append(reales[k]+imaginarios[i]*1j) #Agregar a las columnas los numeros complejos
    m.append(b) #Agregar las columnas a cada fila
```

```python
#Matriz que guarda el Llamado de la funcion vectorizar con la matriz construida
vectorizado=funcionVectorizada(m,0.00000000001)
```

```python
#Matriz que guarda la conversion de imaginarios a angulos
angulo=np.angle(vectorizado)
```

```python
#Comando para generar el fractal
imshow(angulo)
#Comando para exportar el fractal
plt.savefig("fractal_lr.png")
```
