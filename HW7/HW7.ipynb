```python
def funcion(z):
    y=z**5+z**2
    return y

def derivada(z):
    dy=5*(z**4)+2*z
    return dy
```

```python
import numpy as np
```

```python
def NR(x, epsilon):
    root = x
    iteration = 0.
    i=abs(funcion(root))
    while i >= epsilon:
        
        if derivada(root)==0:
            print "error"
            i=epsilon+1
        else:
            iteration += 1 
            root = root - funcion(root)/(derivada(root))
           
            if iteration>=50:
                i=epsilon+1
            else:
                i=abs(funcion(root))
    return root
```

```python
print NR(-5j,0.0000000000000001)
```

```python
vecfun=np.vectorize(NR)

```

```python

a=[]
d=np.linspace(-1,1,200)
for i in range(0,200):
    b=[]
    c=np.linspace(1,-1,200)
    for k in range(0,200):
        
        b.append(d[k]+c[i]*1j)
    a.append(b)

```

```python
vectorizado=vecfun(a,0.00000000001)
print vectorizado
```
