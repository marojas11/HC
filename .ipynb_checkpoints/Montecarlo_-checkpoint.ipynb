```python
x = np.random.rand(10000000)*2 - 1
y = np.random.rand(10000000)*2 - 1
```

```python
index = np.where(x**2+y**2<1)
```

```python
xc = x[index]
yc = y[index]
```

```python
figure(figsize=(9,9))
scatter(x,y, c='k', alpha=0.3)
scatter(xc,yc, c='r', alpha=0.4, s=40)
xlim(-1, 1)
ylim(-1, 1)
```

```python
pi = 4* float(len(xc))/float(len(x))
print ("pi = " '%.16f' % pi)
```
