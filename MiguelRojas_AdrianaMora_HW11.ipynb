```python
#se Importo numpy y matplotlib al igual que shuffle
from random import uniform, shuffle
import matplotlib.pyplot as plt
import numpy as np
   
```

```python
plt.figure(figsize=(20,20))
#Se inicializo X,Y y a
#X, Y siendo 1000 ya que la matriz es 1000 x 1000 al igual que 1000 etapas
#a es la matriz inicializada con todo 0, todavia no hay carros
X = 1000 
Y = 1000 
a = np.zeros(shape=(Y,X))
#Se inicializo una funcion en lo cual entra una lista en el parametro y esta lista se baraja con shuffle y se retorna la lista
def revolver(lista):
        random.shuffle(lista)
        return lista
#Se inincializo una funcion con parametro k siendo cuando carros estaran en la matriz    
def func(k):
    #Se inicializa las variables 
    #vmax la velocidad maxima
    vmax = 5
    #La probabilidad
    p = 1/3.0
    #El contador de distancia
    b=0
    #El numero de carros en la matriz
    carros= k
    #Se creo la lista con el numero de carros buscados
    ini = [0] * carros + [-1] * (X -carros )
    #Se barajo la lista inicial
    hh=[revolver(ini)]
    
    #un recorrido de rango 1000
    for i in range(Y):
        #Se inicializo dos variables a y d.A siendo la que esta antes y D la despues 
        A,D = hh[-1],[-1] * X
       
         #Un recorrido de rango 1000 adentro de un recorrido
        for x in range(X):
            #si A[x] mayor a -1 entonces cumple la condicion y entra al if
            if A[x] > -1:
                #definimos una varibale como Ax
                vi = A[x]
                #Se inicializa d =1 
                d = 1
                #Se hace un recorrido que estara dando mientras cumpla la siguiente condicion
                #Mientras A en la posicion dad sea menor a 0
                while A[(x + d) % X] < 0:
                    d += 1
                #Se crea una variable temporal de la velocidad 
                vtemp = min(vi+1, d - 1, vmax) 
                #Se define la velocidad maxima
                v = max(vtemp - 1, 0) if uniform(0,1) < p else vtemp 
                #Se suman las velocidades para calcular la distancia recorrida
                b+=v
                #Se define D igual a la velocidad en la posicion dad
                D[(x + v) % X] = v 
                
         #se agrega D a la lista de hh
        hh.append(D)
 
 
    #Se crea un recorrido de rango 1000
    for i in range(X):
        # Adentro del recorrido se pone otro recorrido de rango 1000
        for j in range(Y):
            #Se define a en las posiciones i y j
            a[j,i] = 1 if hh[j][i] > -1 else 0
    #Se retorna la matriz a y la distancia b
    return a,b
#Se meten 150 carros en la función    
h,b=func(150)
#Se Corre la grafica.
plt.imshow(h, cmap="Greys", interpolation="nearest")
plt.show()
```

```python
#Se crean dos listas y un contador
y=[]
ii=[]
cont=50
#El contador suma de 10 en 10 en un rango de 0,70
for i in range (0,55):
    
    cont+=10
    #Se pone en la funcion de 50 a 600 
    h,b=func(cont)
    y.append(b)
    ii.append(cont)

plot(ii,y)
```
