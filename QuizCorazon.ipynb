```python
def corazon(x):
    ar=(1-(abs(x)-1)**(2))**0.5
    ab=np.arccos(1-abs(x))-np.pi
    return ar,ab
```

```python
min_x = -2.0
max_x = 2.0
x = linspace(min_x,max_x,10000)

y1,y2 =corazon(x)
plot(x,y2)
plot(x,y1)

```

```python
min_y = amin(y2)
max_y = amax(y1)

n_random = 100000
random_x = random.rand(n_random) * (max_x - min_x) + min_x
random_y = random.rand(n_random) * (max_y - min_y) + min_y
b,c=corazon(random_x)
a=[ (b- random_y > 0.0) & (c-random_y<0.0 )]
index = where(a[0])

scatter(random_x[index], random_y[index], alpha=0.5, c='r')

ylim(-3, 1.1)
```

```python
interval_integral = (max_y-min_y) * (max_x - min_x)
integral  = interval_integral * (size(index)/(1.0*size(random_y)))
print ('dentro:', size(index), 'total:', size(random_y))
print('Valor de la integral:', integral)
```
