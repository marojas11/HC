```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
ano,mes,minas,area = np.genfromtxt("desminadohumanitario.csv", delimiter=",", skip_header=1, unpack=True)
```

```python
f=[]
l=len(ano)

for i in range(0,l):
    
    a=datetime.date(int (ano[i]),int (mes[i]),1)
    f.append(a)
```

```python
plt.scatter(f,minas, c='k', alpha=0.7, s=50)
xlabel("$\mathrm{x}$", fontsize=25)
ylabel("$\mathrm{y}$", fontsize=25)
title("$\mathrm{Datos\ reales}$", fontsize=25)
```
